## Gráficas de dispersión.

```python
# Grafiquemos el primer par de variables.
x = df["sepal length (cm)"]
y = df["sepal width (cm)"]

fig = plt.figure()
plt.scatter(x, y)
plt.show()
```

```python
# Modifiquemos nuestra gráfica anterior.
x = df["sepal length (cm)"]
y = df["sepal width (cm)"]
species = df["target"]

fig = plt.figure(figsize=(6, 5)) # Ajustamos el tamaño.
plt.scatter(x, y, s = 80, c = species) # Graficamos pero con un color para cada tipo.
plt.grid(True) # Muestra la cuadricula.
plt.xlabel("sepal length (cm)")
plt.ylabel("sepal width (cm)")

plt.show()
```

```python
# Modifiquemos nuestra gráfica anterior.
x = "sepal length (cm)"
y = "sepal width (cm)"
species = "target"
fig = plt.figure(figsize=(6,5))

groups = df.groupby(species)
for group in groups: 
    plt.plot(group[1][x], group[1][y], marker='o', linestyle='', markersize=10, label=group[1][species].unique())

plt.grid(True) # Muestra la cuadricula.
plt.xlabel("sepal length (cm)")
plt.ylabel("sepal width (cm)")
plt.legend()
plt.show()
```

```python
features = len(data.feature_names)
print(features)
```

```python
for i in range(features):
    for j in range(features):
        
        indice_x = i
        indice_y = j

        x = df.iloc[:, indice_x]
        y = df.iloc[:, indice_y]
        species = df.iloc[:, 4]

        fig = plt.figure(figsize=(6, 5)) # Ajustamos el tamaño.
        plt.scatter(x, y, s = 80, c = species) # Graficamos pero con un color para cada tipo.
        plt.grid(True) # Muestra la cuadricula.
        plt.xlabel(data.feature_names[indice_x])
        plt.ylabel(data.feature_names[indice_y])
        
        title = str(data.feature_names[indice_x]) + " vs " + str(data.feature_names[indice_y])
        plt.title(title)
        plt.show()
        
        print("")
```

# Jointplot

```python
fig = plt.figure()
sns.jointplot("sepal length (cm)", "sepal width (cm)", data=df, kind='hex')
plt.show()
```

```python
for i in range(features):
    for j in range(features):
        # Obtenemos la primer característica que vamos a graficar.
        feat1 = data.feature_names[i]
        # Obtenemos la segunda característica que vamos a graficar.
        feat2 = data.feature_names[j]
        fig = plt.figure()
        sns.jointplot(feat1, feat2, data=df, kind='hex')
        plt.show()
        print("")
```

```python
setosa = df[df['target'] == 0]
setosa.head()
```

```python
setosa.describe()
```

```python
setosa = setosa.drop(['target'], axis =1)
setosa.head()
```

```python
versicolor = df[df['target'] == 1]
versicolor = versicolor.drop(['target'], axis =1)
versicolor.head()
```

```python
virginica = df[df['target'] == 2]
virginica = virginica.drop(['target'], axis =1)
virginica.head()
```

```python
fig = plt.figure()
plt.hist(setosa['sepal length (cm)'], alpha = 0.4, label = "Setosa")
plt.legend()
plt.show()
```

```python
fig = plt.figure()
plt.hist(setosa['sepal length (cm)'], alpha = 0.4, label = "Setosa")
plt.hist(versicolor['sepal length (cm)'], alpha = 0.4, label = "Versicolor")
plt.hist(virginica['sepal length (cm)'], alpha = 0.4, label = "Virginica")
plt.title("sepal length (cm)")
plt.legend()
plt.show()
```

```python
feat = data.feature_names[0]

fig = plt.figure()
plt.hist(setosa[feat], alpha = 0.4, label = "Setosa")
plt.hist(versicolor[feat], alpha = 0.4, label = "Versicolor")
plt.hist(virginica[feat], alpha = 0.4, label = "Virginica")
plt.title(feat)
plt.legend()
plt.show()
```

```python
for i in range(features):
    feat = data.feature_names[i]

    fig = plt.figure()
    plt.hist(setosa[feat], alpha = 0.4, label = "Setosa")
    plt.hist(versicolor[feat], alpha = 0.4, label = "Versicolor")
    plt.hist(virginica[feat], alpha = 0.4, label = "Virginica")
    plt.title(feat)
    plt.legend()
    plt.show()
    print()
```
